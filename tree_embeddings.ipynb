```python
#import libraries
import os
import xml.etree.ElementTree as ET
```

```python
from treelib import Node, Tree

def add_tree_nodes(tree, parent_name, parent_dim, child_names, child_dims, counter):
    tree.show()
    # root node name
    #in_names = []
    #in_dims = []
    
    # iterates the child nodes
    child_dims_unmodified = child_dims
    for node_i in range(len(child_dims_unmodified)):
        print(len(child_dims_unmodified))
        print(len(child_dims))
        print(node_i)
        if child_dims_unmodified[node_i] in child_dims:
            # get the dimensions and name of a child node
            child_dim = child_dims_unmodified[node_i]
            child_x1, child_x2, child_y1, child_y2 = child_dim
            parent_x1, parent_x2, parent_y1, parent_y2 = parent_dim

            child_name = child_names[node_i]

            # checks if the node is inside the parent one and adds it to the tree if true, after it iterates the child nodes
            if child_x1 >= parent_x1 and child_x2 <= parent_x2 and child_y1 >= parent_y1 and child_y2 <= parent_y2:
                child_name = str(counter) + "_" + child_name
                counter += 1
                tree.create_node(child_name, child_name, parent = parent_name)

                child_names.pop(node_i)
                child_dims.pop(node_i)

                #in_names.append(child_name)
                #in_dims.append(child_dim)

                # gets the modified tree, and the dimensions and names of the nodes that didnt belong to the parent node
                tree, child_names, child_dims = add_tree_nodes(tree, child_name, child_dim, child_names, child_dims, counter)
    return (tree, child_names, child_dims)

def get_tree(name, size):
    tree = Tree()
    tree.create_node("0_root", "0_root")
    
    root_dim = [0, 1500, 0, 2600]
    return add_tree_nodes(tree, "0_root", root_dim, name, size, 1)
```

```python
files = os.listdir('./xml-full/')
for file in files:
    print(file)
    tree = ET.parse(os.path.join('./xml-full/',file))
    root = tree.getroot()
    object = root.findall("object")
    name = [x.findtext("name") for x in object]
    size = [[int(x.findtext("bndbox/xmin")), int(x.findtext("bndbox/xmax")), int(x.findtext("bndbox/ymin")),\
             int(x.findtext("bndbox/ymax"))] for x in object]
    
    file_tree = get_tree(name, size)
    
    break
```
